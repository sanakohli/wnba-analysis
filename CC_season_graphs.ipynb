```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
CC_db = pd.read_csv('CC_Game_Stats.csv')
```

```python
# reversing order to be chronological through season
CC_db = CC_db.iloc[::-1].reset_index(drop=True)
```

```python
CC_db.head()
```

```python
# creating index column
CC_db['index'] = CC_db.index
```

```python
# plotting assists per game
plt.figure(figsize=(10, 6))
plt.plot(CC_db['index'], CC_db['AST'], marker='o', linestyle='-')

# annotate each game with a 'W' or 'L' based on game result
for i, row in CC_db.iterrows():
    plt.annotate(
        'W' if row['W/L'] == 'W' else 'L', 
        (row['index'], row['AST']),
        textcoords="offset points",
        xytext=(0, -20),  # Adjust the offset as needed
        ha='center'
    )

# adding labels
plt.title('Caitlin Clark\'s Assists Per Game')
plt.xlabel('Game #')
plt.ylabel('Assists')

plt.grid(True)
plt.show()
```
